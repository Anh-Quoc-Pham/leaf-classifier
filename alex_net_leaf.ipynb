```python
# ============================================================
#  ALEXNET TRANSFER LEARNING (FREEZE FEATURES) CHO PHÂN LOẠI 12 LOÀI LÁ
#  - Root: /kaggle/input/more-plant-health
#  - Bên trong: "A Database of Leaf Images Practice towards Plant Conservation with Plant Pathology"
#  - Mỗi loài: species/healthy/*.JPG, species/diseased/*.JPG
#  - AlexNet pretrained ImageNet, freeze features, train FC
#  - Log đầy đủ: train/val loss & acc mỗi epoch
#  - Cuối cùng: classification report + confusion matrix
# ============================================================

import os
from pathlib import Path

import numpy as np
import pandas as pd
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

import torchvision.transforms as T
import torchvision.models as models

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

import matplotlib.pyplot as plt
from tqdm import tqdm
import random

# ---------------------------
# 0. FIX SEED CHO TÁI LẬP
# ---------------------------
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(42)

# ---------------------------
# 1. CẤU HÌNH ĐƯỜNG DẪN DATA
# ---------------------------
ROOT = Path("/kaggle/input/more-plant-health")
DATA_DIR = ROOT / "A Database of Leaf Images Practice towards Plant Conservation with Plant Pathology"

print("DATA_DIR:", DATA_DIR)
print("Tồn tại?:", DATA_DIR.exists())

if not DATA_DIR.exists():
    raise FileNotFoundError("Không tìm thấy DATA_DIR, hãy kiểm tra lại đường dẫn.")

# ---------------------------
# 2. QUÉT ẢNH & TẠO DATAFRAME
#    - Mỗi thư mục con trong DATA_DIR là 1 loài
#    - Trong đó có healthy/ và diseased/ (trừ 2 trường hợp đặc biệt)
#    - Ảnh đuôi .JPG (viết hoa)
#    - Nhãn là LOÀI (không phân biệt healthy/diseased)
# ---------------------------
species_dirs = sorted([d for d in DATA_DIR.iterdir() if d.is_dir()])
print("Số loài (species folders):", len(species_dirs))
print("Các loài:", [d.name for d in species_dirs])

if len(species_dirs) == 0:
    raise RuntimeError("Không tìm thấy thư mục loài nào trong DATA_DIR.")

species_to_idx = {d.name: idx for idx, d in enumerate(species_dirs)}
idx_to_species = {v: k for k, v in species_to_idx.items()}

image_paths = []
labels = []
species_names = []

for species_dir in species_dirs:
    species_name = species_dir.name
    label = species_to_idx[species_name]

    # Lần lượt vào healthy/ và diseased/
    for sub in ["healthy", "diseased"]:
        sub_dir = species_dir / sub
        if not sub_dir.exists():
            continue

        # Ảnh đuôi .JPG (viết hoa)
        for img_file in sub_dir.glob("*.JPG"):
            image_paths.append(str(img_file))
            labels.append(label)
            species_names.append(species_name)

df = pd.DataFrame({
    "path": image_paths,
    "label": labels,
    "species": species_names,
})

print("Tổng số ảnh thu được:", len(df))
print(df.head())
print("\nPhân bố theo loài:")
print(df["species"].value_counts())

if len(df) == 0:
    raise RuntimeError("DataFrame rỗng (len(df) == 0). Có thể không có file .JPG nào theo cấu trúc đã mô tả.")

# ---------------------------
# 3. CHIA TRAIN / VAL (STRATIFIED)
# ---------------------------
train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["label"],
    random_state=42,
)

print("\nSố ảnh train:", len(train_df))
print("Số ảnh val  :", len(val_df))

# ---------------------------
# 4. TRANSFORMS
#    - Resize về 224x224 (đúng với AlexNet)
#    - Augment nhẹ cho train
#    - Chuẩn hóa theo mean/std ImageNet
# ---------------------------
IMG_SIZE = 224

train_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.RandomHorizontalFlip(),
    T.RandomRotation(15),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]),
])

val_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]),
])

# ---------------------------
# 5. CUSTOM DATASET + DATALOADER
# ---------------------------
class LeafSpeciesDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df.reset_index(drop=True)
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = row["path"]
        label = int(row["label"])

        img = Image.open(img_path).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, label

BATCH_SIZE = 64  # AlexNet paper dùng 128, ở đây dùng 64 cho GPU Kaggle

train_dataset = LeafSpeciesDataset(train_df, transform=train_transform)
val_dataset   = LeafSpeciesDataset(val_df,   transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,  num_workers=2)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

print("\nDataLoader sẵn sàng!")

# ---------------------------
# 6. ALEXNET PRETRAINED + FREEZE FEATURES
# ---------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

num_classes = len(species_to_idx)

alexnet = models.alexnet(weights=models.AlexNet_Weights.IMAGENET1K_V1)

# Thay lớp FC cuối cùng (4096 -> num_classes)
in_features = alexnet.classifier[6].in_features
alexnet.classifier[6] = nn.Linear(in_features, num_classes)

alexnet = alexnet.to(device)

# Freeze toàn bộ feature extractor (các lớp conv)
for param in alexnet.features.parameters():
    param.requires_grad = False

print("\nModel AlexNet (đã thay fc cuối & freeze features):")
print(alexnet)

# ---------------------------
# 7. LOSS, OPTIMIZER, SCHEDULER
# ---------------------------
criterion = nn.CrossEntropyLoss()

optimizer = optim.SGD(
    filter(lambda p: p.requires_grad, alexnet.parameters()),
    lr=0.001,          # 0.01 như paper là cho từ-scratch, ở đây TL nên dùng 0.001 ổn hơn
    momentum=0.9,
    weight_decay=5e-4,
)

scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)

# ---------------------------
# 8. HÀM TRAIN / EVAL
# ---------------------------
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss, correct, total = 0.0, 0, 0
    
    for imgs, labels in tqdm(loader, desc="Training", leave=False):
        imgs, labels = imgs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * imgs.size(0)
        _, preds = outputs.max(1)
        correct += preds.eq(labels).sum().item()
        total += labels.size(0)

    return running_loss / total, correct / total


def evaluate(model, loader, criterion, device):
    model.eval()
    running_loss, correct, total = 0.0, 0, 0

    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc="Validating", leave=False):
            imgs, labels = imgs.to(device), labels.to(device)

            outputs = model(imgs)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * imgs.size(0)
            _, preds = outputs.max(1)
            correct += preds.eq(labels).sum().item()
            total += labels.size(0)

    return running_loss / total, correct / total

# ---------------------------
# 9. TRAIN N EPOCHS & LƯU MODEL TỐT NHẤT
# ---------------------------
EPOCHS = 20
best_acc = 0.0
best_wts = None

history = {
    "train_loss": [],
    "train_acc":  [],
    "val_loss":   [],
    "val_acc":    [],
}

for epoch in range(EPOCHS):
    print(f"\n===== Epoch {epoch+1}/{EPOCHS} =====")
    
    train_loss, train_acc = train_one_epoch(alexnet, train_loader, criterion, optimizer, device)
    val_loss,   val_acc   = evaluate(alexnet, val_loader, criterion, device)

    scheduler.step()

    history["train_loss"].append(train_loss)
    history["train_acc"].append(train_acc)
    history["val_loss"].append(val_loss)
    history["val_acc"].append(val_acc)

    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"Val   Loss: {val_loss:.4f} | Val   Acc: {val_acc:.4f}")

    if val_acc > best_acc:
        best_acc = val_acc
        best_wts = alexnet.state_dict().copy()

print("\nBest validation accuracy:", best_acc)

if best_wts is not None:
    alexnet.load_state_dict(best_wts)
    torch.save(alexnet.state_dict(), "alexnet_leaf_species_freeze_best.pth")
    print("Đã lưu model tốt nhất vào alexnet_leaf_species_freeze_best.pth")
else:
    print("Không có best_wts, có thể training có vấn đề.")

# ---------------------------
# 10. BÁO CÁO CUỐI: CLASSIFICATION REPORT + CONFUSION MATRIX
# ---------------------------
alexnet.eval()
all_labels = []
all_preds  = []

with torch.no_grad():
    for imgs, labels in tqdm(val_loader, desc="Predicting on val", leave=False):
        imgs = imgs.to(device)
        outputs = alexnet(imgs)
        _, preds = outputs.max(1)

        all_labels.extend(labels.numpy())          # labels đang ở CPU
        all_preds.extend(preds.cpu().numpy())      # preds từ GPU về CPU

all_labels = np.array(all_labels)
all_preds  = np.array(all_preds)

target_names = [idx_to_species[i] for i in range(num_classes)]

print("\n=== CLASSIFICATION REPORT (VAL SET) ===")
print(classification_report(all_labels, all_preds, target_names=target_names, digits=4))

cm = confusion_matrix(all_labels, all_preds)
print("Confusion matrix (raw counts):")
print(cm)

# Vẽ confusion matrix
fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(cm, interpolation='nearest')
ax.figure.colorbar(im, ax=ax)
ax.set(
    xticks=np.arange(cm.shape[1]),
    yticks=np.arange(cm.shape[0]),
    xticklabels=target_names,
    yticklabels=target_names,
    ylabel='Nhãn thật (True label)',
    xlabel='Nhãn dự đoán (Predicted label)',
    title='Confusion Matrix - AlexNet (Frozen Features)'
)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

thresh = cm.max() / 2.0
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, format(cm[i, j], "d"),
                ha="center", va="center",
                color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import torch

def predict_single_image(img_path, model, transform, device, idx_to_species):
    # Load ảnh
    img = Image.open(img_path).convert("RGB")

    # Áp dụng transform giống val_transform
    x = transform(img).unsqueeze(0).to(device)

    # Dự đoán
    model.eval()
    with torch.no_grad():
        logits = model(x)
        probs = torch.softmax(logits, dim=1)[0]
        pred_idx = torch.argmax(probs).item()

    species_name = idx_to_species[pred_idx]
    confidence = float(probs[pred_idx])

    print("Ảnh:", img_path)
    print("Dự đoán:", species_name)
    print("Độ tin cậy:", f"{confidence:.4f}")

    return species_name, confidence
```
